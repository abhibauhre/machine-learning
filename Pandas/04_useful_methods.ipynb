```python
def quality(total_bill,tip):
    if tip / total_bill > 0.35:
        return 'Generous'
    else :
        return 'other'
```

```python
quality(0.24,5)
```

```python
df['Quality'] =df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis = 1)#axis = 1 its work on row-wise
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])#it syntax is easier than upper block function
#it also do same and output is same
```

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv("D:\UNZIP-FOR-NOTEBOOKS-FINAL\03-Pandas\tips.csv")

def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
  
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
  
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```
