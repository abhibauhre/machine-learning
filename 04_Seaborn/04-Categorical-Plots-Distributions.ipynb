```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data = df)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y = 'math score',data = df,hue= 'gender')

#optional move the legend outside 
plt.legend(bbox_to_anchor = (1.05,1),loc = 2 ,borderaxespad = 0.)
```

```python
# NOTICE HOW WE HAVE TO SWITCH X AND Y FOR THE ORIENTATION TO MAKE SENSE!
sns.boxplot(x='math score',y='parental level of education',data=df,orient='h')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender',width=0.3)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y= 'math score',data = df)
```

```python
## Advanced Plots

#We can use a boxenplot and swarmplot to achieve the same effect as the boxplot
#and violinplot, but with slightly more information included. Be careful when using these plots, 
#as they often require you to educate the viewer with how the plot is actually constructed.
#Only use these if you are sure your audience will understand the visualization.
```

```python
df.head()
```

```python
sns.swarmplot(x='math score',data = df)
```

```python
sns.swarmplot(x='math score',data=df,size=2)
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df,size=3)
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df,size=3)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',y='math score',data=df,hue='gender')
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(x='race/ethnicity',y='math score',data=df,hue='gender',dodge=True)
```

```python
### boxenplot (letter-value plot)

# Official Paper on this plot: https://vita.had.co.nz/papers/letter-value-plot.html

# This style of plot was originally named a letter value plot
# because it shows a large number of quantiles that are defined as letter values. 
# It is similar to a box plot in plotting a nonparametric representation of a distribution 
# in which all features correspond to actual observations.
#  By plotting more quantiles, it provides more information about the shape of the distribution, 
# particularly in the tail
```

```python
sns.boxenplot(x='math score',y='race/ethnicity',data=df,palette='Dark2')
```

```python
sns.boxenplot(x='race/ethnicity',y='math score',data=df,palette='Dark2')
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='race/ethnicity',y='math score',data=df,hue='gender',palette='Dark2')
```
