```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dm_office_sales.csv")
df
```

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

```python
plt.figure(figsize=(12,3) , dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df,hue = 'level of education') #(hue) that uses different colors to represent categorical variables in your plots, adding an extra dimension of information without requiring additional axes.
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

```python
plt.figure(figsize=(12,3) , dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df,hue = 'level of education',palette = 'Dark2')
#https://matplotlib.org/stable/users/explain/colors/colormaps.html(this is for colour)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200) # change the marker size 
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)#alpha parameter is used to operate the capacity of plots
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
```

```python
plt.figure(figsize=(12,8))
# Sometimes its nice to do BOTH hue and style off the same column
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```
